```python
import pandas as pd

df = pd.read_excel('State_to_State_Migration_Table_2023_T13.xlsx', sheet_name='Table', header=None)

header_row = df.iloc[:, 1:].map(lambda x: 'alabama' in str(x).lower()).any(axis=1).idxmax()
state_headers = df.iloc[header_row].ffill()

texas_row = df[df[0].astype(str).str.strip().str.lower() == 'texas'].index[0]
texas_values = df.iloc[texas_row]

texas_df = pd.DataFrame({
    'Residence 1 Year Ago': state_headers,
    'Number of Migrants': pd.to_numeric(texas_values, errors='coerce')
}).reset_index(drop=True)

texas_filtered = texas_df.iloc[1::2].dropna(subset=['Number of Migrants'])

texas_filtered = texas_filtered.dropna(subset=['Residence 1 Year Ago'])

invalid_labels = ['total', 'texas', 'united states']
texas_states_only = texas_filtered[~texas_filtered['Residence 1 Year Ago'].astype(str).str.strip().str.lower().isin(invalid_labels)]

top_10_texas = texas_states_only.nlargest(10, 'Number of Migrants')

top_10_texas.to_csv('texas_inflows_2023.csv', index=False)
display(top_10_texas)
```

```python
import pandas as pd
import numpy as np
import glob
import os

files = glob.glob(os.path.expanduser('~/Downloads/*State_to_State*2022*.xls*'))
if not files:
    raise FileNotFoundError("Could not find the official State_to_State Excel file for 2022 in your Downloads.")

df = pd.read_excel(files[0], header=None)

state_headers = df.iloc[6].astype(str).str.strip().tolist()
texas_row_idx = df[df[0].astype(str).str.strip() == 'Texas'].index

if len(texas_row_idx) == 0:
    raise ValueError("Could not locate the 'Texas' data row in this spreadsheet.")

texas_data = df.iloc[texas_row_idx[0]].tolist()
texas_migration = pd.DataFrame({
    'Residence 1 Year Ago': state_headers,
    'Amount of People': texas_data
})

texas_migration['Amount of People'] = texas_migration['Amount of People'].astype(str).str.replace(',', '', regex=False)
texas_migration['Amount of People'] = pd.to_numeric(texas_migration['Amount of People'], errors='coerce')
texas_migration = texas_migration.dropna()

invalid_labels = ['Total', 'United States', 'Texas', 'Puerto Rico', 'Elsewhere', 'nan']
texas_migration = texas_migration[~texas_migration['Residence 1 Year Ago'].str.contains('|'.join(invalid_labels), case=False, na=False)]

top_sources_to_texas = texas_migration.nlargest(10, 'Amount of People')

display(top_sources_to_texas)
top_sources_to_texas.to_csv('final_texas_inflows_2022.csv', index=False)
```

```python
import pandas as pd
import numpy as np

# 1. Load the Excel file completely raw
df = pd.read_excel('~/Downloads/State_to_State_Migrations_Table_2021.xls', header=None)

# 2. Extract the column headers (Row 6 contains the origin state names)
# Extract the Texas data row (Find where Column 0 is exactly 'Texas')
state_headers = df.iloc[6].astype(str).str.strip().tolist()
texas_row_idx = df[df[0].astype(str).str.strip() == 'Texas'].index[0]
texas_data = df.iloc[texas_row_idx].tolist()

# 3. Create a clean DataFrame from these paired rows
texas_migration = pd.DataFrame({
    'Residence 1 Year Ago': state_headers,
    'Number of Migrants': texas_data
})

# 4. Clean commas, convert to numbers, and drop missing values
texas_migration['Number of Migrants'] = texas_migration['Number of Migrants'].astype(str).str.replace(',', '', regex=False)
texas_migration['Number of Migrants'] = pd.to_numeric(texas_migration['Number of Migrants'], errors='coerce')
texas_migration = texas_migration.dropna()

# 5. Filter out totals, duplicates, and non-states
invalid_labels = ['Total', 'United States', 'Texas', 'Puerto Rico', 'nan']
texas_migration = texas_migration[~texas_migration['Residence 1 Year Ago'].str.contains('|'.join(invalid_labels), case=False, na=False)]

# 6. Get the true top 10 largest inflows to Texas
top_sources_to_texas = texas_migration.nlargest(10, 'Number of Migrants')

# 7. Display and save clean data
display(top_sources_to_texas)
top_sources_to_texas.to_csv('texas_inflows_2021.csv', index=False)



```

```python
# 2019 data loading
import pandas as pd

# 1. Load the ENTIRE sheet with no special column rules to avoid crashes
df_raw = pd.read_excel('State_to_State_Migrations_Table_2019.xls', sheet_name='Table', skiprows=4)

# 2. Slice the columns: grab column L (index 11) all the way to the very end of the sheet
df = df_raw.iloc[:, 11:].copy()

# 3. Build a clean list of column names step-by-step
clean_columns = []
current_state = ""

# Loop through every column one by one
for position in range(len(df.columns)):
    state_box = str(df.iloc[0, position])
    metric_box = str(df.iloc[1, position])
    
    if state_box != 'nan':
        current_state = state_box
        
    clean_columns.append(f"{current_state}_{metric_box}")

# Assign names to columns
df.columns = clean_columns

# Rename the first column to our standard name
df.rename(columns={df.columns[0]: 'Current_Residence'}, inplace=True)

# Throw away header rows used for titles
df = df.iloc[3:].reset_index(drop=True)
pd.set_option('display.max_rows', None)

print("Data loaded successfully! Columns are ready.")

```

```python
import pandas as pd

# 1. Find the physical row index where the first column is 'Texas'
texas_row_idx = df[df.iloc[:, 0].astype(str).str.strip().str.lower() == 'texas'].index

if texas_row_idx.empty:
    print("Could not find the Texas row in the dataset.")
else:
    # 2. Get all columns that contain the actual Estimates (ignoring the '_nan' Margin of Error columns)
    estimate_cols = [c for c in df.columns if not c.lower().endswith('_nan') and c != 'Current_Residence']
    
    # 3. Extract the Texas row for those estimate columns
    texas_data = df.loc[texas_row_idx, estimate_cols]
    
    # 4. Flip (transpose) the row into a vertical table
    texas_2019 = texas_data.transpose().reset_index()
    texas_2019.columns = ['Residence 1 Year Ago', 'Number of Migrants']
    
    # 5. Clean up the state names (removes the long prefix, leaves just 'California', 'Florida', etc.)
    texas_2019['Residence 1 Year Ago'] = texas_2019['Residence 1 Year Ago'].astype(str).apply(lambda x: x.split('_')[-1].strip())
    
    # 6. Convert metrics to pure numbers
    texas_2019['Number of Migrants'] = pd.to_numeric(texas_2019['Number of Migrants'], errors='coerce')
    
    # 7. Drop rows with missing values or redundant labels
    texas_2019 = texas_2019.dropna(subset=['Residence 1 Year Ago', 'Number of Migrants'])
    invalid_labels = ['Total', 'United States', 'United States2', 'Texas', 'Puerto Rico']
    texas_2019 = texas_2019[~texas_2019['Residence 1 Year Ago'].isin(invalid_labels)]
    
    # 8. Get the true top 10 largest inflows
    top_sources_to_texas = texas_2019.nlargest(10, 'Number of Migrants')
    
    # 9. Display and save
    display(top_sources_to_texas)
    top_sources_to_texas.to_csv('texas_inflows_2019.csv', index=False)


```
